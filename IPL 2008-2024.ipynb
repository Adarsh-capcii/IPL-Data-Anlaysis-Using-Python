# Data Cleaning

```python
[i[0] for i in match_f[match_f['city'].isna()]['venue'].value_counts().reset_index()['index'].str.split()]
```

```python
# Define a function to extract the city name based on the venue if city is missing
def fill_city(row):
    if pd.isna(row['city']):
        # Get the most frequent first word in 'venue' names where 'city' is NaN
        most_frequent_city = match_f[match_f['city'].isna()]['venue'].mode().iloc[0].split()[0]
        return most_frequent_city
    return row['city']

# Apply the function to the 'city' column
match_f['city'] = match_f.apply(fill_city, axis=1)
match_f['city'].isnull().sum()
```

**Here by value_counts of unique value i.e. Dubai stadium and sharjah stadium and it is 33 and 18 respectively can be replaced as per starting name that whose name starts with dubai replace with dubai and which name starts with sarjah replace it with sarjah**

# calculate strike rate season wise

```python
# Group by season and player, summing runs and balls faced
season_player_stats = ipl.groupby(['season', 'batter']).agg({
    'batsman_runs': 'sum',
    'ball': 'count'
}).reset_index()
season_player_stats.head()
```

```python
# Calculate strike rate
season_player_stats['strike_rate'] = ((season_player_stats['batsman_runs'] / season_player_stats['ball']) * 100).round(2)
season_player_stats.sort_values(by='strike_rate',ascending=False).head()
```

```python
# Sort by season and strike rate (descending)
season_player_stats = season_player_stats.sort_values(['season', 'strike_rate'], ascending=[True, False])

# Display top players in each season
top_strike_rates = season_player_stats.groupby('season').head(1)  # Top player per season
print(top_strike_rates)
```

```python
season_player_runs=ipl.groupby(['season','batter','batting_team'])['batsman_runs'].sum().reset_index()
season_player_runs
# Sort the data by season and runs
season_player_runs = season_player_runs.sort_values(['season', 'batsman_runs'], ascending=[True, False])

# Get the top scorer for each season
top_scorers = season_player_runs.groupby('season').head(1).reset_index(drop=True)
top_scorers
```

```python
top_batter_season=pd.merge(season_player_stats,top_scorers,on=['season','batter'],how='right')
top_batter_season=top_batter_season.drop(columns="batsman_runs_y")
top_batter_season=top_batter_season.rename(columns={"batsman_runs_x":"Runs",
                                 "season":"Season",
                                 "batter":"Player",
                                 "ball":"Balls",
                                 "batting_team":"Team"})
top_batter_season
```

```python
plt.figure(figsize=(18, 6))
sns.barplot(data=top_batter_season, x='Season', y='strike_rate', hue='Player', dodge=False)
plt.title('Orange Cap holder Strike Rates in Each Season')
plt.xlabel('Season')
plt.ylabel('Strike Rate')
plt.legend(title='player', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
